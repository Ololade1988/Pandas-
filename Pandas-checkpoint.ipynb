```python
# what is series and how to create 
### Series is one dimensional array. i.e one column and multiplle rows. 
### it can be created from multiple objects e.g tuple, int, list etec.
```

```python
name =["sancho","white","rice","stone","onana","rashford"]
series1=pd.Series(name)
series1
```

```python
# To create series from tuple
tup1=("one","two","three","four")
sr2=pd.Series(tup1)
sr2
```

```python
# To create series from dict
dic={"name":"sancho","age":12,"salary":8373}
pd.Series(dic)
```

```python
# To create series with customise index
name =["sancho","white","rice","stone","onana","rashford"]
series1=pd.Series(name, index=["Y","W","E","R","T","D"])
series1
```

```python
# How to convert series to dataframe
series1.to_frame(name="people's name")
```
